Complete the exercises below to solidify your knowledge and understanding of unsupervised learning model evaluation.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets

data = datasets.load_wine()

X = pd.DataFrame(data["data"], columns=data["feature_names"])
y = pd.Series(data["target"])
```

## 1. Train a KMeans clustering model on the data set using 8 clusters and compute the silhouette score for the model.

```python
from sklearn import cluster
kmeans=cluster.KMeans(n_clusters=8)
kmeans.fit(X)
y_pred=kmeans.predict(X)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(X, y_pred)
```

## 2. Train a KMeans clustering model on the data set using 5 clusters and compute the silhouette score for the model.

```python
kmeans5=cluster.KMeans(n_clusters=5)
kmeans5.fit(X)
y5=kmeans5.predict(X)
```

```python
silhouette_score(X, y5)
```

## 3. Train a KMeans clustering model on the data set using 3 clusters and compute the silhouette score for the model.

```python
kmeans3=cluster.KMeans(n_clusters=3)
kmeans3.fit(X)
y3=kmeans3.predict(X)
```

```python
silhouette_score(X, y3)
```

## 4. Use elbow curve visualizations to see if you can determine the best number of clusters to use.

The Yellowbrick library has 3 metrics that you can plot using the `metric` parameter:

- **distortion**: mean sum of squared distances to centers
- **silhouette**: mean ratio of intra-cluster and nearest-cluster distance
- **calinski_harabaz**: ratio of within to between cluster dispersion

```python
#conda install -c districtdatalabs yellowbrick
```

```python
from yellowbrick.cluster import KElbowVisualizer
```

```python
modelo=cluster.KMeans()

visual=KElbowVisualizer(modelo, k=(1, 15))
visual.fit(X)

visual.poof();
```

## 5. Try performing the same elbow tests with an AgglomerativeClustering model and compare the results you get to the KMeans results.

```python
#Con Agglomaritve Clustering
aggclus8=cluster.AgglomerativeClustering(n_clusters=8, linkage='single')
aggclus8.fit(X)

y_ac8=aggclus8.labels_

#Prueba silhouette
silhouette_score(X, y_ac8)
```

```python
aggclus3=cluster.AgglomerativeClustering(n_clusters=3, linkage='single')
aggclus3.fit(X)

y_ac3=aggclus3.labels_

#Prueba silhouette
silhouette_score(X, y_ac3)
```

```python
aggclus5=cluster.AgglomerativeClustering(n_clusters=5, linkage='single')
aggclus5.fit(X)

y_ac5=aggclus5.labels_

#Prueba silhouette
silhouette_score(X, y_ac5)
```

```python
modelo=cluster.AgglomerativeClustering()

visual=KElbowVisualizer(modelo, k=(1, 15))
visual.fit(X)

visual.poof();
```

En el primer caso, con K-means los valores obtenidos con la prueba de silhouette score indican que la clusterización es mejor que la de Agglomerative Clustering. 

De acuerdo a la gráficas obtenida, en ambos casos el número óptimo de clusters es de 3. 

## 6. Create and plot a scatter matrix showing how the clusters are grouped across all the different combinations of variables in the data.

Use the model and number of clusters that returned the best result above.

```python
plt.figure(figsize=(15,15))
pd.plotting.scatter_matrix(X[['alcohol','malic_acid','ash','alcalinity_of_ash']],c=y3,cmap='viridis');
```

En el scatterplot se pueden observar los clusters de cada relación de columnas. Se seleccionó el KMeans con 3 clusters para observar los grupos.

## 7. Apply a PCA transform and plot the first two principle components with the plot point colors determined by cluster.

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA().fit(X)

pca_x=pca.transform(X)
```

```python

fig=plt.figure(figsize=(12, 8))

plt.scatter(pca_x[:,0], pca_x[:,1],c=y3,cmap='viridis')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')

y=np.ones(X.shape[0])

plt.show();
```

## 8. Generate a series of t-SNE plots showing the clusters at a variety of perplexities.

```python
from sklearn.manifold import TSNE
```

```python
fig, ax = plt.subplots(1, 4, figsize=(15, 7))

for i,p in enumerate([10, 20, 30, 40]):
    
    tsne=TSNE(perplexity=p)
    
    x_emb=tsne.fit_transform(X)
    
    ax[i].scatter(x_emb[:,0], x_emb[:,1], c=y3, cmap='viridis')
    ax[i].set_title('Perplexity={}'.format(p))
    
plt.show();
```
