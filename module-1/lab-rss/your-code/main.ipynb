### 1. Use feedparser to parse the following RSS feed URL.

```python
url = 'http://feeds.feedburner.com/oreilly/radar/atom'
```

```python
rssfeeds = feedparser.parse(url)
```

### 2. Obtain a list of components (keys) that are available for this feed.

```python
rssfeeds.keys()
```

### 3. Obtain a list of components (keys) that are available for the *feed* component of this RSS feed.

```python
rssfeeds['feed'].keys()
```

### 4. Extract and print the feed title, subtitle, author, and link.

```python
print('Título: {}'.format(rssfeeds['feed']['title']))
print('Subtítulo: {}'.format(rssfeeds['feed']['subtitle']))
print('Link: {}'.format(rssfeeds['feed']['link']))
```

### 5. Count the number of entries that are contained in this RSS feed.

```python
len(rssfeeds['entries'])
```

### 6. Obtain a list of components (keys) available for an entry.

*Hint: Remember to index first before requesting the keys*

```python
rssfeeds['entries'][0].keys()
```

### 7. Extract a list of entry titles.

```python
titulos=[]
for i in range (len(rssfeeds['entries'])):
    titulos.append(rssfeeds['entries'][i]['title'])
titulos
```

### 8. Calculate the percentage of "Four short links" entry titles.

```python
contador=0
for i in titulos:
    if i.startswith('Four'):
        contador+=1
total=len(titulos)
porcentaje=(contador*total)/100
print("{}%".format(porcentaje))
```

### 9. Create a Pandas data frame from the feed's entries.

```python
import pandas as pd
```

```python
df = pd.DataFrame(rssfeeds['entries'])
df.head()
```

### 10. Count the number of entries per author and sort them in descending order.

```python
authors = df.groupby('author', as_index=False).agg({'title':'count'})
authors.columns = ['author', 'entries']
authors.sort_values('entries', ascending=False)
```

### 11. Add a new column to the data frame that contains the length (number of characters) of each entry title. Return a data frame that contains the title, author, and title length of each entry in descending order (longest title length at the top).

```python
df['length_title']=df['title'].apply(len)
new_data=df[['title', 'author', 'length_title']]
new_data.sort_values('length_title', ascending=False)
new_data
```

### 12. Create a list of entry titles whose summary includes the phrase "machine learning."

```python

import re
'''Estructura de re.search'
re.search('lo que quiero encontrar', frase donde buscaré)
si no encuentra en la frase, me devuelve None
    '''
titulos2=[]

for i in range (len(rssfeeds['entries'])):
    aux=rssfeeds['entries'][i]['summary']
    if re.search('machine learning',aux)!=None:
        titulos2.append(rssfeeds['entries'][i]['title'])
titulos2
```
