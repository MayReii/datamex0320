#### Use the appropriate plot to visualize the column `Gender`. There are 2 ways of doing it. Do it both ways.

*Hint: Use matplotlib and seaborn*

```python
# your code here- 1st way
generos=titanic['PassengerId'].groupby(titanic['Gender']).count().reset_index()

generos.plot(kind='bar');
plt.xticks(list(range(2)),['Female','Male'], rotation=55)
plt.xlabel('Gender')
plt.ylabel('Count')
```

```python
# your code here- 2nd way
sns.countplot(x='Gender',data=titanic);
```

#### Use the appropriate plot to visualize the column `Pclass` 


```python
# your code here
pclass=titanic['Pclass'].value_counts()

pclass.plot(kind='bar',color=['red','yellow','blue']);
plt.xlabel('Clase')
plt.ylabel('Count')
```
